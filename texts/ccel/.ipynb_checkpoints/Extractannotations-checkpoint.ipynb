```python
import re

inlineanns = 0

chapternotecombos = {}
chapternotelist = {}
chapternotetargets = {}
chapternotesentid = {}
for i in range(0,72):
    #chapternotelist.append([])
    chapternotelist['chap' + str(i)] = []
    
chaptertarget2sentid = {}
    
sups = 1
def getanns(fname):
    global inlineanns
    global sups
    f = open(fname + '.htm')
    workfname = 'work/' + fname + '-work.htm'
    outf = open(workfname,'w')
    for l in f:
        l = re.sub('<p><b id=','<p id=',l)
        l = re.sub('<[Bb]\s+(id="[^"]+">[^<]+)</[Bb]>','<sup \g<1></sup>',l)
        l = re.sub('<P id="','<p id="',l)

        l = re.sub('<p><small class="ok" id(=[^>]+>[^<]+)</small>','<p id\g<1>',l)
        l = re.sub('<p><a name="','<p id="',l)
        l = re.sub('<small class="ok" id','<p id',l)
        l = re.sub('(</sup>)\s\n','\g<1> ',l)
        l = re.sub('(</small>\s*<br>)','\g<1>\n',l)
        l = re.sub('<[/]*[Bb]>','',l)
        l = re.sub('(A\.\s*D\.)\s*\n','\g<1> ',l)
        print(l,end='',file=outf)
            
    f.close()
    outf.close()
    
    curid = ''
    curngram = ''
    word = '\s*[^\s]+'
    ngram = '(' + word+word+word+word + ')'
    f = open(workfname)
    for l in f:
        l = re.sub('<em[^>]*>','',l)
       # if(re.search('A.D. 1370-1383',l)):
        #    print('test',l)
        m = re.search('<p id="([^"]+)"',l)
        if(1 and m):
            curid = m[1]
            curngram = ''
            curlabel = ''
            savel = l
            l = re.sub('<[^>]+>','',l)
            l = re.sub('[\s\n]+$','',l)
            l = re.sub('^[\s\n]+','',l)
            curlabel = l
            #print('1',fname,curid,curlabel,savel,sep='--')
            #if(re.search('A.D. 1370-1383',l)):
            #print('test2',curlabel,curid,l,sep='--')
            continue
        
        cleanw = '[A-Za-z\-\']+[;:,!]*\s+'
        m = re.search('('+cleanw+cleanw+cleanw+cleanw+cleanw+cleanw+cleanw+')',l)
        if(1 and m and curid and not curngram):
            curngram = re.sub('[\(\)]','.',m[1])
            curngram = re.sub('^\s*','',curngram)
            curngram = re.sub('\s$','',curngram)
            #print('label',fname,curid,curngram,curlabel,sep='\t')
            chapternotelist[fname].append([curlabel,curngram])
            chapternotecombos[fname+'\t'+curngram] = 0
            chapternotetargets[fname+'\t'+curngram] =  curid
            curid = ''

        if(re.search('<[Bb]\s+(id="[^"]+">[^<]+)</[Bb]>',l)):
            print(l)
        l = re.sub('<sup class="ok">','<sup>',l)
        l = re.sub('<a name="([^"]+)"><sup>','<sup id="\g<1>">',l)
        l = re.sub('</sup>\s*</a>','</sup>',l)
        l = re.sub('</SUP>','</sup>',l)
        l = re.sub('<sup><[Ii] [Ii]d="([^"]+)">','<sup id="\g<1>">',l)
        l = re.sub('<sup\s+Id','<sup id',l)
        l = re.sub('(<sup[^>]*>)\s*<[Ii]>','\g<1>',l)
        l = re.sub('</[Ii]>\s*</sup>','</sup>',l)
        l = re.sub('(<sup id="[^"]+")[^>]+>','\g<1>>',l)
        l = re.sub('<sup><b id','<sup id',l)
        l = re.sub('(sup[^>]*>[^<]+)<[Ii]>','\g<1>',l)
        l = re.sub('(sup[^>]*>[^<]+)</[Ii]>','\g<1>',l)
        l = re.sub('&#8212;','-',l)
        l = re.sub('<sup\s+>','<sup>',l)
        
        while(re.search('<sup\s+id="([^"]+)">([^<]+)</sup>' + ngram,l)):
            m = re.search('<sup\s+id="([^"]+)">([^<]+)</sup>' + ngram,l)
            if(m):
                inlineanns = inlineanns + 1
                curtarget = m[1]

                #print('<sup>',inlineanns,fname,m[1],m[2],l,sep='\t',end='')
                matchs = re.sub('\s+',' ',m[3])
                matchs = re.sub('^\s+','',matchs)
                tmpkey = fname+'\t'+matchs
                label = re.sub('^\s+','',m[2])
                label = re.sub('\s$','',label)
                chapternotelist[fname].append([label,matchs])
                chapternotecombos[tmpkey] = 0
                chapternotetargets[tmpkey] =  curtarget
                #print('got1','matchs='+matchs,label,tmpkey,chapternotetargets[tmpkey],sep='--')
                
            else:
                print('nongram',fname,l)
            l = re.sub('<sup\s+[Ii]d="([^"]+)">([^<]+)</sup>' + ngram,'',l,1)

        l = re.sub('<sup>\s*<[iI]>','<sup>',l)
        l = re.sub('</[iI]>\s*</sup>','</sup>',l)
        l = re.sub('(</sup>)([^ ])','\g<1> \g<2>',l)
        while(re.search('<sup>([^<]+)</sup>' + ngram,l)):
            m = re.search('<sup>([^<]+)</sup>' + ngram,l)
            if(m):
                inlineanns = inlineanns + 1
                curtarget = 'notarg'

            #print(inlineanns,fname,m[1],m[2],l,sep='\t',end='')
                matchs = re.sub('\s+',' ',m[2])
                matchs = re.sub('^\s+','',matchs)
                label = re.sub('^\s+','',m[1])
                label = re.sub('\s$','',label)
                chapternotelist[fname].append([label,matchs])
                chapternotecombos[fname+'\t'+matchs] = 0
                chapternotetargets[fname+'\t'+matchs] = curtarget
                #print('got2','matchs='+matchs,label,tmpkey,chapternotetargets[tmpkey],sep='--')
                
            else:
                print('nongram',fname,l)
            l = re.sub('<sup>([^<]+)</sup>' + ngram,'',l,1)

        if(re.search('<sup',l)):
            print(sups,fname,'sup',l)
            sups = sups + 1
        while(not re.search('<p ',l) and re.search('\(A\.D\.\s+([0-9\-]+)\)'+ ngram,l)):

            m = re.search('\(A\s*\.D[\.,]\s*([0-9\-]+)\)'+ ngram,l)
            if(m):
                inlineanns = inlineanns + 1
                curtarget = m[2]
                curdate = m[1]
                print(fname,curdate,curtarget,l)
            l = re.sub('\(A\s*\.D[\.,]\s*([0-9\-]+)\)'+ ngram,'',l,1)
            
            
    f.close()

for i in range(1,72):
    getanns('chap'+ str(i))

f = open('../bury-all.xml')
for l in f:
    m = re.search('^<s xml:id="(ch[0-9]+)([^"]+)',l)
    if(m):
        cursentid = m[1] + m[2]
        curchap = re.sub('^ch[0]*','',m[1])
        curkey = 'chap' + curchap
        if(curkey in chapternotelist):
            #if(re.search('ch01pa16se007',l)):
                #print('checking',curkey,l)
            for foo in chapternotelist[curkey]:
                if(re.search('[\(\)]',foo[1], re.IGNORECASE)):
                    print('badstring',curkey,foo[1])
                    continue

                #if(re.search(foo[1],l, re.IGNORECASE)):
                searchs = foo[1]
                searchs = re.sub('<.+','',searchs)
                searchs = re.sub('<[Ii]>','<hi rend="italic">',searchs)
                searchs = re.sub('</[Ii]>','</hi>',searchs)
                if(re.search('\s([A-Z][A-Z][A-Z]+)',searchs)):
                    searchs = re.sub('\s([A-Z][A-Z][A-Z]+)',' <hi rend="small-caps">\g<1></hi>',searchs)
                    #print(searchs,l)
                searchs = re.sub('^([AI]\s+)(<hi rend="small-caps">)','\g<2>\g<1>',searchs)
                searchs = re.sub('(\s+|,\s+)','[ ,]+',searchs)
                if(0 and re.search('apparent tranquillity of Gaul',foo[1])):
                    print(searchs,l)
                searchs = re.sub('([,;]\s*|\s*)$','[;,]*',searchs)

                if( re.search('^([A-Z][A-Z]+|[IA]\s+[A-Z][A-Z]+)',searchs) and not re.search('^[IV][IV]',searchs)):
                    searchs = re.sub('^([A-Z][A-Z]+|[IA]\s+[A-Z][A-Z]+)','<hi rend="small-caps">\g<1></hi>',searchs)

                searchl = re.sub('<ref[^>]+>','',l)
                searchl = re.sub('<pb[^>]+>','',searchl)

                searchl = re.sub('\s+',' ',searchl)
                if(re.search(searchs,searchl,re.IGNORECASE)):
                    #print(curkey,foo[0],foo[1],l)
                    if(chapternotecombos[curkey+'\t'+foo[1]]):
                        print('doubling',curkey+'\t'+foo[1],l)
                    chapternotelist[curkey+'\t'+foo[1]] = foo[0]
                    chapternotesentid[curkey+'\t'+foo[1]] = cursentid
                    chapternotecombos[curkey+'\t'+foo[1]] = chapternotecombos[curkey+'\t'+foo[1]] + 1
                #else:
                    #testpat = 'ch02pa11se001'
                    #if(re.search(testpat,l)):
                        #print('failed',searchs,testpat,l,sep='-')

f.close()

hits = 1
outf = open('marginalia.tsv','w')

ccel2persid = {}
persid2marg = {}
for foo in chapternotecombos:
    if(not chapternotecombos[foo] == 1):
        print('failed',chapternotecombos[foo],foo)
    else:
        ccel2persid[foo.split('\t')[0] + '\t' + chapternotetargets[foo]] = chapternotesentid[foo]
        #print(chapternotetargets[foo],foo.split('\t')[0],sep='--')
        #print('hit',hits,'chapternotesentid='+chapternotesentid[foo],'chapternotetargets='+chapternotetargets[foo],'chapternotelist='+chapternotelist[foo],foo)
        print(chapternotesentid[foo],chapternotetargets[foo],chapternotelist[foo],foo,sep='\t',file=outf)
        if(chapternotesentid[foo] in persid2marg):
            persid2marg[chapternotesentid[foo]] = persid2marg[chapternotesentid[foo]] + ' ' + chapternotelist[foo]
            print('appended',chapternotesentid[foo],persid2marg[chapternotesentid[foo]])
        else:
            persid2marg[chapternotesentid[foo]] = chapternotelist[foo]
        #print(chapternotesentid[foo],chapternotelist[foo])
        hits = hits + 1
print('hits',hits)

outf.close()
#for foo in chapternotelist:
   # print(foo,chapternotelist[foo])
            
```

```python

import re

chapheadlist = {}

totkeys = 0
def getheads(cnum):
    global totkeys
    fname = 'chap' + str(cnum) + '.htm'
    f = open(fname)
    sawhead = 0
    inhead = 0
    for l in f:
        if(re.search('<p class="cite"',l)):
            sawhead = 1
            inhead = 1
            continue
            #print(fname,l,end='')
        if(inhead and re.search('</[Pp]>',l)):
            inhead = 0
        if(inhead):
            workl = l
            #print('chap' + str(cnum),l,end='')
            curhead = re.sub('\s*\n+$','',l)
            curhead = re.sub('^\s*','',curhead)
            curhead = re.sub('\s*(&#8212;$)','',curhead)
            curhead = re.sub('\s*<br>','',curhead)
            if(re.search('&#8212;',curhead)):
                print('baddel',cnum,curhead)
            chapheadlist[cnum].append(curhead)
            while(re.search('"#([^"]+)',workl)):
                m = re.search('"#([^"]+)',workl)
                curtag = m[1]
                workl = re.sub('"#([^"]+)','',workl,1)
                curkey = re.sub('\.htm$','',fname) + '\t' + curtag
                if(not curkey in ccel2persid):
                    print(curkey,'nopersid',l)
                else:
                    totkeys = totkeys + 1
                    #print(totkeys,curkey,ccel2persid[curkey],l,end='')
            #print(fname,l,end='')
    if(not sawhead):
        print(fname)
    f.close()
    
for i in range(1,72):
    chapheadlist[i] = []

    
for i in range(1,72):
    getheads(i)

for i in range(1,72):
    continue
    for foo in chapheadlist[i]:
        print(i,foo)
    #print(i,chapheadlist[i])
print('done')
```

```python
import re
from fuzzywuzzy import fuzz

def checkrefseq(curl):
    if(not re.search('<list xml:id',l) and not re.search('<ref',l)):
        return
    targval = 0
    prevtargval = 0
    while(re.search('<ref target="ch[0-9]+pa([0-9]+)se([0-9]+)">',curl)):
        m = re.search('<ref target="ch[0-9]+pa([0-9]+)se([0-9]+)">',curl)
        paval = m[1]
        seval = m[2]
        curtag = 'pa' + paval + 'se' + seval
        targval = 100 * int(m[1]) + int(m[2])
        if(targval < prevtargval):
            print('target jump',prevtag,curtag,curl)
        prevtargval = targval
        prevtag = curtag
        curl = re.sub('<ref target="ch[0-9]+pa([0-9]+)se([0-9]+)">','',curl,1)

outf = open('bury-work.xml','w')

f = open('../bury-all.xml')
for l in f:
    m = re.search('chapter"\s+n="([0-9]+)"',l)
    if(m):
        chapnum = int(m[1])
        curchap = re.sub('^([0-9])$','0\g<1>',m[1])
        #print('curchap',curchap)
    m = re.search('<list><item>',l)
    if(m):
        l = re.sub('<list>','<list xml:id="ch'+curchap+'toc">',l)
        l = re.sub('<note[^>]+>([^<]+)</note>','\g<1>',l)
        curtoc = l
        curtocitem = 1
        while(re.search('<item>',l)):
            m = re.search('<item>(.+)</item>',l)
            if(m):
                curhead = m[1]
                curhead = re.sub('</item>.*','',curhead)
            else:
                curhead = 'nohead'
            maxres = 0
            bestm = ''
            for foo in chapheadlist[chapnum]:
                workl = re.sub('<[^>]+>','',foo)
                res = fuzz.ratio(workl.lower(),re.sub('<[^>]+>','',curhead.lower()))
                if(res > maxres):
                    maxres = res
                    bestm = foo
                    c1 = workl.lower()
                    c2 = re.sub('<[^>]+>','',curhead.lower())
            if(maxres < 90):    
                print('fuzz',maxres,c1,c2,chapnum,curhead,bestm,l,sep='--')
                
            workl = bestm
            workl = re.sub('^\s*','',workl)
            workl = re.sub('>([^< ]+)</a>([, ]+[A-Za-z]+)','>\g<1>\g<2></a>',workl)
            #print('newbest',bestm)
            while(maxres > 80 and re.search('"#([^"]+)">([^<]+)',workl)):
                m = re.search('"#([^"]+)">([^<]+)',workl)
                curtag = m[1]
                curkey = 'chap' + str(chapnum) + '\t' + curtag

                curlink = m[2]
                curlink = re.sub('^\s*','',curlink)
                curlink = re.sub('\s*$','',curlink)
                m = re.search('('+curlink+')',curhead, flags=re.IGNORECASE)
                if(not m):
                    print('failed',curlink,l,sep='--')
                    curmatch = 'none'
                else:
                    curmatch = m[1]
                if(not re.search(curmatch,l)):
                    print('curheadfail',curmatch,l,sep='--',end='')
                newhead = re.sub('('+curmatch+')','<ref target="'+ccel2persid[curkey]+'">\g<1></ref>',curhead,1)
                l = re.sub(curhead,newhead,l)
                oldhead = curhead
                curhead = newhead
                #print(chapnum,bestm,workl,curlink,oldhead,curhead,l,sep='--')
                workl = re.sub('"#([^"]+)">([^<]+)','',workl,1)
                if(not curkey in ccel2persid):
                    print(curkey,'nopersid',l)
                #else:
               #     print(totkeys,curkey,curlink,ccel2persid[curkey],l,end='')

            curlab = 'ch' + curchap + 'toc'+re.sub('^([0-9])$','0\g<1>',str(curtocitem))
            l = re.sub('<item>','<item xml:id="'+curlab+'">',l,1)
            curtocitem = curtocitem + 1
    l = re.sub('(</ref>)([^<]+)(</item>)','\g<2>\g<1>\g<3>',l)
    l = re.sub('(<item[^>]*>)([^<]+)(<ref[^>]*>)','\g<1>\g<3>\g<2>',l)
    l = re.sub('(\s+and)(</ref>)','\g<2>\g<1>',l)
    checkrefseq(l)
    m = re.search('<s xml:id="([^"]+)',l)
    if(m):
        sentid = m[1]
        if(sentid in persid2marg):
            print('<note type="margin" xml:id="'+sentid+'marg">'+persid2marg[sentid]+'</note>',file=outf)
    print(l,file=outf,end='')
        
f.close()
outf.close()
```
