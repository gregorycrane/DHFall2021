```python
import re

ammianpat = '(Ammianus|Ammian\.)([ ,\()]+)([xvil]+)\.[ ]+([0-9, ]+)'
def getcits(fname):
    
    f = open(fname)
    
    for l in f:
        m = re.search(ammianpat,l)
        if(m):
            curchaps = re.sub(',\s*$','',m[4])
            curchaps = re.sub(',\s+','-',curchaps)
            print('Ammianus ' + m[3] + ' ' + curchaps)
    f.close()

getcits('bury-nohyph-04.xml')
```
