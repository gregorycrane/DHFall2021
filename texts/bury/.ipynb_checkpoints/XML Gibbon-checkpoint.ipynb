![sample table of contents page](./gibbon-bury-01-00x.jpg)

Readers can scan the extracted text and then jump to the relevant pages but the brief summaries of events have lost their precise link to the text they describe. Not all marginalia accompany the start of a new paragraph and simple techniques cannot restore all the text from the marginalia back to their place in the narrative.

By placing the marginalia into the table of contents, however, the isolated marginalia can now become formed into a timeline, allowing readers to skim a chapter watching events unfold over time or searching for events at or after a particular date.

In a digital edition, we want to be able to support both modalities -- quickly scanned tables of contents and embedded summaries that readers can scan as they move quickly through the text. The best approach is the keep the link between summary marginalia and the text, generating the tables of contents on the fly as desires. 

The Online Library of Liberty did not transcribe the detailed tables of contents at all. Using OCR to transcribe the tables of contents would not be difficult but the transcribed table of contents cannot (as noted above) be automatically reintegrted into the running text.  

Transcribing marginalia with OCR, however, is quite onerous because it is usually necessary to draw boxes around the marginalia and the main text in such a way that the marginalia can be inserted into the running text. Even then, marking the marginalia as marginalia requires another step that may not lend itself to automation. 

The digitized version of the Decline and Fall posted by the Our Decline site (http://www.ourdecline.com/smartboard/shop/gibbone/rome/volume1/cntnt6.htm) manages to combine both modalities. First, it provides a table of contents that _only_ includes events associated with dates:

![Our decline TOC](./gibbon-ourdecline-chap06-toc.jpg)

Then it preserves the marginalia in their proper places in the text:

![Our decline TOC](././outdecline-gibbon-06.jpg)

```python
import re

def getdates(suffix):
    #outfname = 'bury-' + suffix + '.xml'
    #outf = open(outfname,'w')

    f = open('bury-nohyph-' + suffix + '.xml')

    adpat = '([0-9]+)([,\s]+)<span class="sc"><span class="entity ad">ad</span></span>'
    adpat2 = '<span class="sc"><span class="entity ad">ad</span></span>\s+([0-9]+)'
    bcpat = '<span class="sc"><span class="entity bc">bc</span></span>\s+([0-9]+)'
    bcpat2 = '([0-9]+)\s+<span class="sc"><span class="entity bc">bc</span></span>'

    blankpage = '<span class="pb"><span class="decoration">Edition: current; Page: </span><span class="bracket">\[</span>(none|[a-z])<span class="bracket">\]</span></span>'
    nodiv = 1
    infig = 0
    inpoem = 0
    incit = 0
    usepa = "1"
    lastnote = ''
    firstchapter = 1
    curline = 0
    print('<milestone unit="volume" n="'+ suffix + '"/>',file=outf)
    for l in f:
        
        l = re.sub(blankpage,' ',l)
        if( re.search('<ul class="poem">',l)):
            inpoem = 1
            l = '<quote n="poem">'
        if(inpoem and re.search('<li ',l)):
            curline = curline + 1
            l = re.sub('<li[^>]+>(.+)</li>','<l n="' + suffix + '.' + str(curline)+ '">\g<1></l>',l)
        if(re.search('</ul>',l) and inpoem):
            l = re.sub('</ul>','</quote>',l)
            inpoem = 0
            
        if(nodiv):
            l = re.sub('</div>','',l)
        if(re.search('subtype="chapter"',l) and  nodiv == 0):
            print('</div>',file=outf)
            usepa = "1"
        
        if(re.search('<p [^>]+pa1">([^<]+</item>)',l)):
            usepa = "2"
        l = re.sub('<p [^>]+pa1">([^<]+</item>)','<list><item>\g<1>',l)
        l = re.sub('(<item>[^<]+)</p>','\g<1></item></list>',l)

        if(re.search('ch[0-9]+\.pa' + usepa + '"',l)):
            print('<div type="textpart" subtype="chaptertext">',file=outf)
        if(re.search('<div',l)):
            nodiv = 0
        l = re.sub('xml:target','target',l)
        l = re.sub('<div ([^>]*\s*)class=','<div \g<1>type="textpart" subtype=',l)
        l = re.sub('lighttp','light',l)
        l = re.sub('</span>([,;:\."!])</span>','</span></span>\g<1>',l)
        l = re.sub('<div[^>]+(figcaption[^>]+>)([^<]+)</div>','<figDesc>\g<2></figDesc>',l)

        while(re.search(adpat,l)):
            l = re.sub(adpat,'<date when="\g<1>">\g<1>\g<2>A.D.</date>',l,1)
        while( re.search(bcpat,l)):
            l = re.sub(bcpat,'<date when="-\g<1>">B.C. \g<1></date>',l,1)
        while( re.search(adpat2,l)):
            l = re.sub(adpat2,'<date when="-\g<1>">A.D. \g<1></date>',l,1)
        while(re.search(bcpat2,l)):
            l = re.sub(bcpat2,'<date when="-\g<1>">\g<1> B.C.</date>',l,1)
        l = re.sub('when="-([0-9])"','when="-000\g<1>"',l)
        l = re.sub('when="-([0-9][0-9])"','when="-00\g<1>"',l)
        l = re.sub('when="-([0-9][0-9][0-9])"','when="-0\g<1>"',l)
        l = re.sub('when="([0-9])"','when="000\g<1>"',l)
        l = re.sub('when="([0-9][0-9])"','when="00\g<1>"',l)
        l = re.sub('when="([0-9][0-9][0-9])"','when="0\g<1>"',l)
        l = re.sub('<span class="(csc|sc)">([^<]+)</span>','<hi rend="smallcap">\g<2></hi>',l)
        l = re.sub('<span class="sc"><span class="entity bc">bc</span></span>','<date>B.C.</date>',l)        

        if( re.search('subtype="figure"',l)):
            infig = 1
            l = re.sub('<div[^>]+([0-9]+_figure_[0-9]+)[^>]+>','<figure xml:id="v'+ suffix + '-fig\g<1>">',l)
        l = re.sub('<img src="([^"]+)[^>]+>','<graphic url="\g<1>"/>',l)
        l = re.sub('\s+id="',' xml:id="',l)
        if(infig and re.search('</div>',l)):
            l = re.sub('</div>','</figure>',l)
            infig = 0
        l = re.sub('<sup>([^<]+)</sup>','<hi rend="super">\g<1></hi>',l)
        l = re.sub('<pb n="','<pb n="v'+ suffix +'p',l)
        l = re.sub('<div type="textpart" subtype="bibl">([^<]+)(</note>)','</quote>\n<bibl>\g<1></bibl>\g<2>',l)
        if(re.search('<li[^>]+><div type="textpart" subtype="closer">',l)):
            continue
        if(re.search('^<div type="textpart" subtype="closer">',l)):
            continue
        l = re.sub('<p class="','<p rend="',l)
        
        if( re.search('<div type="textpart" subtype="cit">',l)):
            incit = 1
            l = re.sub('<div type="textpart" subtype=','<quote n=',l)
        if( re.search('</div>',l) and incit):
            l = re.sub('</div>','</quote>',l)
            incit = 0
            
        l = re.sub('<span class="roman">([^<]+)</span>','<num type="roman">\g<1></num>',l)
        
        m = re.search('id="(ch[0-9]+n[0-9]+)"',l)
        if(m):
            lastnote = m[1]
        m = re.search('>([0-9]+[abc])\s*</a>',l)
        if(m):
            notenum = m[1]
            newnote = re.sub('(ch[0-9]+n)','\g<1>' + notenum,lastnote)
            newpar = newnote + 'pa1'
            l = re.sub('<div[^>]+><a[^>]+>[^<]+</a><p[^>]+>','<note xml:id="' +newnote + '"><p xml:id="' +newpar + '">',l)
            #print(m[1],lastnote,l)
        l = re.sub('<span xml:id="[^>]+>([^<]+)</span>','\g<1>',l)
        print(l,end='',file=outf)
        
        
    f.close()
    print('\n</div>',file=outf)

outfname = 'bury-' + 'all' + '.xml'
outf = open(outfname,'w')
f = open('gibbonheader.xml')
for l in f:
    print(l,end='',file=outf)
f.close()
    
getdates('01')
getdates('02')
getdates('03')
getdates('04')
getdates('05')
getdates('06')
getdates('07')
getdates('08')
getdates('09')
getdates('10')
getdates('11')
getdates('12')
print('</body>\n</text>\n</TEI>',file=outf)

outf.close()
```

```python
import re

def countit(pat,line):
    rval = 0
    words = line.split()
    for foo in words:
        if(foo == pat):
            rval = rval + 1
            
    return rval

def getgreek(fname):
    f = open(fname)
    for l in f:
        ingreek = 0
        if(re.search('[α-ωΑ-Ω]x',l)):
           print(l)
        words = l.split()
        i = 0
        prevfoo = ''
        prevfoo2 = ''
        for foo in words:
            if(re.search('[α-ωΑ-Ω]',foo)):
                if( ingreek ):
                    curgreek = curgreek + ' ' + foo
                else:
                    curgreek = foo
                ingreek = 1

            else:
                if(ingreek):
                    if(not countit(prevfoo,l) == 1 or 1):
                        print(countit(prevfoo,l),pregram, '<foreign xml:lang="grc">'+curgreek+"</foreign>",l)

                if( i > 1):
                    pregram = words[i-2] + ' ' + words[i-1] + ' ' + words[i]
                if( i == 1):
                    pregram = words[0]
                ingreek = 0
            prevfoo2 = prevfoo
            prevfoo = foo
            i = i + 1
    
    f.close()
    print('done')
    
getgreek('bak/decline-bury-02.txt')
```
