```python
import re

def addtags(taglist,sentence):
    sawkeys = {}
    while(re.search("'([^']+)',\s+'([^']+)'",taglist)):
        m = re.search("'([^']+)',\s+'([^']+)'",taglist)
        if(m):
            searchkey = m[1]
            tag = m[2]
            if(not searchkey in sawkeys):
                sentence = re.sub(searchkey,'<'+tag+'>'+searchkey+'</'+tag+'>',sentence)
            
            sawkeys[searchkey] = 1
            #print(m[1],m[2])
        taglist = re.sub("'([^']+)',\s+'([^']+)'",' ',taglist,1)
    return sentence

for row in rows:
    taggedsent = addtags(row[1],re.sub('\s+',' ',row[0]))
    print(taggedsent)
#addtags("('Parthians', 'NORP'), ('Crassus', 'PERSON'), ('1', 'CARDINAL'),  ('Crassus', 'PERSON')",'')
   
```
